```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### 1. Cálculo del área de la lámina rectangular

```python
x_bounds = [2, 6]
y_bounds = [1, 3]
```

#### Cálculo del área

```python
area = (x_bounds[1] - x_bounds[0]) * (y_bounds[1] - y_bounds[0])
```

#### Graficar la región rectangular

```python
fig, ax = plt.subplots()
rect = plt.Rectangle((x_bounds[0], y_bounds[0]), x_bounds[1] - x_bounds[0], y_bounds[1] - y_bounds[0],
                      edgecolor='blue', facecolor='lightblue', alpha=0.7)
ax.add_patch(rect)
ax.set_xlim(0, 8)
ax.set_ylim(0, 4)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(f'Región rectangular (Área = {area} u²)')
plt.grid()
plt.show()
```

### 2. Cálculo del volumen del tetraedro

###### Definimos la región limitada por los planos x=0, y=0, z=0, x+y+z=1 

###### Función de densidad f(x, y, z) = 5x - 3y

```python
def f(x, y):
    return 5 * x - 3 * y
```

###### Volumen del tetraedro por integración

```python
res = 50 # Resolución para discretizar
```

```python
x = np.linspace(0, 1, res)
y = np.linspace(0, 1, res)
X, Y = np.meshgrid(x, y)
Z = 1 - X - Y
Z[Z < 0] = 0
```

##### Aproximación del volumen como suma de los valores de la función

```python
volumen = np.sum(f(X, Y) * Z) * (1 / res)**2
```

##### Graficar el tetraedro y la función

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, facecolors=plt.cm.viridis(f(X, Y)/np.max(f(X, Y))), alpha=0.7, edgecolor='k')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Tetraedro y función f(x, y, z)')
plt.show()

print(f"Área de la región: {area} u²")
print(f"Volumen del tetraedro: {volumen:.4f} u³")
```
