# Introduccion a Matplotlib

[Matplotlib] es una biblioteca que permite la creacion de figuras y grafios de calidad, mediante el uso de Python
* Permite la creacion de grafios de manera sencilla y eficiente
* Permite la integracion de graficos y figuras de Jupyter Notebook

# Import

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Mostrar los grafios integrados
%matplotlib inline
```

# Representacio Graica de Datos

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

##### Tambien se puede proporcionar dos matrices: una para el eje horisontal y otra para el eje vertical

```python
plt.plot([-3, -1, 0 ,4, 7], [1, 4, 6, 7, 8 ])
plt.axis([-4, 8, 0, 10]) #[xmin, xmax, ymin, ymax]
plt.show()
```

##### Se sigue el mismo procedimiento para pintar una funcion matemaati

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Tambien puede modificarse el estilo de la grafica para que contenga mas informacion

```python
plt.plot(x, y)
plt.title("Square Function")
plt.xlabel("Chido")
plt.ylabel("y = x^2")
plt.grid(True)
plt.show()
```

Pueden superponerse graficas y cambiar el estilo de las lineas

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

Separando en dierentes lineas las funciones

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Function")
plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.xlabel("Chido")
plt.ylabel("y = x^2")
plt.grid(True)
plt.show()
```

Para poder diferenciar entre ambas funciones siempre es reconocible aadir una leyenda

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Function")
plt.plot(x, y, 'b--', label = "x^2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.xlabel("Chido")
plt.ylabel("y = x^2")
plt.grid(True)
plt.legend(loc = "best") #La situa en la mejor ubicacion
plt.show()
```

Tambien puede crearse dos graficas que no se superpongan, estas graficas se organizan en un grid y se denominan subplost

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Function")
plt.subplot(1, 2, 1) # 1 Row, 2 columna, 1st Subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 Row, 2 columna, 2nd Subplot
plt.plot(x, y2, 'g')
plt.xlabel("Chido")
plt.ylabel("y = x^2")
plt.grid(True)
plt.show()
```

Para que las graficas o queden tan ajustadas se puede hacer la figura mas grande

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.figure(figsize = (14, 6))

plt.title("Square Function")
plt.subplot(1, 2, 1) # 1 Row, 2 columna, 1st Subplot
plt.plot(x, y, 'b--')
plt.xlabel("x1", fontsize = 10)
plt.xlabel("y1", fontsize = 10)
plt.grid(True)

plt.subplot(1, 2, 2) # 1 Row, 2 columna, 2nd Subplot
plt.plot(x, y2, 'g')
plt.xlabel("x2", fontsize = 10)
plt.xlabel("y2", fontsize = 10)
plt.xlabel("Chido")
plt.grid(True)
plt.show()
```

## Scatter Plots

```python
from numpy.random import rand
x, y = rand (2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand (2, 100)
x2, y2 = rand (2,100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.title("Sctter")
plt.grid(True)
plt.show()
```

# Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212)
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel('Value')
plt.ylabel('Frecuquency')
plt.show()
```
